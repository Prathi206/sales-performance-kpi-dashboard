```python
import pandas as pd

df = pd.read_csv('superstore_final.csv', encoding='latin1')

df.head()
```

```python
df['Order_Month'] = df['Order Date'].dt.month
df['Order_Year'] = df['Order Date'].dt.year
```

```python
df.head()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], format='mixed', errors='coerce')

df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='mixed', errors='coerce')

df['Order_Month'] = df['Order Date'].dt.month

df['Order_Year'] = df['Order Date'].dt.year

df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)

top_products.plot(kind='bar')

plt.title('Top 10 Products by Sales')

plt.xlabel('Product Name')

plt.ylabel('Sales')

plt.xticks(rotation=75)

plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum()

region_sales.plot(kind='bar')

plt.title('Sales by Region')

plt.xlabel('Region')

plt.ylabel('Sales')

plt.show()
```

```python
monthly_sales = df.groupby('Order_Month')['Sales'].sum()

monthly_sales.plot(kind='line', marker='o')

plt.title('Monthly Sales Trend')

plt.xlabel('Month')

plt.ylabel('Sales')

plt.grid(True)

plt.show()
```

```python
category_sales = df.groupby('Category')['Sales'].sum()

category_sales.plot(kind='pie', autopct='%1.1f%%')

plt.title('Category Wise Sales')

plt.ylabel('')

plt.show()
```

```python
print("Business Insights")

print("\n1. Region with Highest Sales:")
print(df.groupby('Region')['Sales'].sum().sort_values(ascending=False).head(1))

print("\n2. Top Selling Category:")
print(df.groupby('Category')['Sales'].sum().sort_values(ascending=False).head(1))

print("\n3. Top Selling Product:")
print(df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(1))

print("\n4. Month with Highest Sales:")
print(df.groupby('Order_Month')['Sales'].sum().sort_values(ascending=False).head(1))
```
